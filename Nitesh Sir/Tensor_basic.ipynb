#Creating Tensor

```python
#Using empty
a =torch.empty(2 ,3)
print (a, a.dtype, a.shape)
```

```python
#Using Zeros
torch.zeros(3,4)
```

```python
#Using Ones
torch.ones(4,5)
```

```python
#Using rand ->random
torch.rand(2,2)
```

```python
#Use of seed
torch.manual_seed(50)
torch.rand(2,2)
```

```python
# Using tensor - custom tensor
torch.tensor([[1,2,3,5],[4,5,6,0]])
```

```python
#Other ways to create tensor

#arange
print("Using arrange ->", torch.arange(0,10,2))

#Using linspace 
print("Using linspace ->", torch.linspace(0,15,3))

#Using eye
print("Using eye ->", torch.eye(5)) #create square matrix

#Using full
print("Using full ->", torch.full((2,3),5)) # all values are 5 in tensor of size 2X3
```
